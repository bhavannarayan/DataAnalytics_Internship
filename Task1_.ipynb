```python
import pandas as pd

df = pd.read_csv('/content/KaggleV2-May-2016.csv.zip')  # replace with your actual file
```

```python
df.head()          # Preview the top rows
df.info()          # Data types and null counts
df.describe()      # Summary statistics for numerical data
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df['age'].fillna(df['age'].mean(), inplace=True)

```

```python
df.isnull().sum()        # Find missing values per column

# Options:
df.dropna()              # Remove rows with missing values
df.fillna(0)             # Fill with 0
df['age'].fillna(df['age'].mean(), inplace=True)  # Fill age with mean
```

```python
print(df.columns)
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df = df.drop_duplicates()
```

```python
df['gender'] = df['gender'].str.strip().str.lower().replace({'male': 'Male', 'm': 'Male', 'f': 'Female', 'female': 'Female'})
```

```python
# Convert to datetime
df['appointmentday'] = pd.to_datetime(df['appointmentday'])

# Format to dd-mm-yyyy string
df['appointmentday'] = df['appointmentday'].dt.strftime('%d-%m-%Y')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['age'] = df['age'].astype(int)
df['scholarship'] = df['scholarship'].astype(bool)
df['hipertension'] = df['hipertension'].astype(bool)
df['diabetes'] = df['diabetes'].astype(bool)       # age to integer
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
